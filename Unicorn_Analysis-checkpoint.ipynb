```python
#import libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#read data
data_base = pd.read_csv("unicorns_2022.csv")

#check dimensions
data_base.shape
```

```python
#first records
data_base.head()
```

```python
#columns
data_base.columns
```

```python
#check information type
data_base.info()
```

```python
#null fields
data_base.isnull().sum()
```

```python
#grafic
plt.figure(figsize = (15,6))
plt.title("analysing null data")
sns.heatmap(data_base.isnull(), cbar = False);
```

```python
#unique fields
data_base.nunique()
```

```python
#unique values - Rank1
data_base["Industry"].value_counts()
```

```python
#unique values - Rank2
industry_analyse = round(data_base["Industry"].value_counts(normalize = True) * 100, 2)
```

```python
industry_analyse.head(10)
```

```python
plt.figure( figsize=(15,6) )
plt.title("Industry Analyse")
plt.bar( industry_analyse.index[0:10], industry_analyse.head(10))
plt.xticks( rotation=45, ha='right' );
```

```python
country_analyse = round(data_base["Country"].value_counts(normalize = True) * 100, 2)
```

```python
plt.figure(figsize = (15, 6))
plt.title("Analyse the countries that generate unicorns - Top 5")
plt.pie(
    country_analyse.head(5),
    labels = country_analyse.index[0:5],
    shadow = True,
    startangle = 90,
    autopct = "%1.2f%%");
```

```python
#conversion to date
data_base["Date Joined"] = pd.to_datetime(data_base["Date Joined"])
data_base["Date Joined"].head()
```

```python
#year and month
data_base["Month"] = pd.DatetimeIndex(data_base["Date Joined"]).month
data_base["Year"] = pd.DatetimeIndex(data_base["Date Joined"]).year

data_base.head()
```

```python
 #analytical table 
cluster_analysis = data_base.groupby(by = ["Country", "Year", "Month", "Company"]).count()["Industry"].reset_index()
```

```python
cluster_analysis.loc[
    cluster_analysis["Country"] == "United States"
]
```

```python
#transforming the value column
data_base["Valuation ($B)"] = data_base["Valuation ($B)"].replace("$", "")

data_base.head()
```

```python
country_analysis = data_base.groupby(by = ["Country"]).sum()["Valuation ($B)"].reset_index()

country_analysis.head()
```

```python
valuation_analyse = country_analysis.sort_values("Valuation ($B)", ascending = False)
valuation_analyse.head()
```

```python
plt.figure(figsize = (15,6))
plt.title("Valuation - Country")
plt.plot(valuation_analyse["Country"], valuation_analysis["Valuation ($B)"])
plt.xticks(rotation = 45, ha = "right");
```

```python
plt.figure( figsize=(15,6) )
plt.title("Valuation Analyse - Top 10")
plt.bar( valuation_analyse["Country"].head(10), valuation_analyse["Valuation ($B)"].head(10))
plt.xticks( rotation=45, ha='right' );
```

```python
plt.figure( figsize=(15,6) )
plt.title("Valuation Analyse - Top 5")
plt.bar( valuation_analyse["Country"].head(5), valuation_analyse["Valuation ($B)"].head(5))
plt.xticks( rotation=45, ha='right' );
```
